```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/covid_19_data.csv")
data.head()
```

```python
data.info()
```

```python
data['ObservationDate']=data['ObservationDate'].astype('datetime64[ns]')
data['Confirmed']=data['Confirmed'].astype('int64')
data['Deaths']=data['Deaths'].astype('int64')
data['Recovered']=data['Recovered'].astype('int64')
data.info()
```

```python
data.head()
```

```python
### Active Cases
data['Active']=data['Confirmed']-data['Deaths']-data['Recovered']
data.head(50)
data.iloc[50:100]
```

```python
### last/currennt  date
data['ObservationDate'].max()
```

```python
## Fetch the Latest date data
latest_data=data[data['ObservationDate']==data['ObservationDate'].max()]
latest_data.head()
```

```python
latest_data.shape
```

```python
latest_data['Country/Region'].value_counts()
```

```python
latest_data['Country/Region'].nunique()
```

```python
latest_data['Country/Region'].unique()
```

```python
#Count of cases of each country
countires=latest_data.groupby('Country/Region')[['Confirmed','Deaths','Recovered','Active']].sum()
countires=countires.reset_index()
countires
```

```python
#Cases in India
countires[countires['Country/Region']=="India"]
```

```python
### Plot on World map
import plotly.express as px
world_map=px.choropleth(countires,locations='Country/Region',locationmode='country names',color='Confirmed',hover_name='Country/Region',title='Confirmed Cases')
world_map.show()
```

```python
## Top contries having maximum confirmed and recovered cases
top=latest_data.groupby('Country/Region')[['Confirmed','Recovered']].sum().reset_index()
top=top.sort_values(['Confirmed'],ascending=False)
top.head()
```

```python
### India
india=data[data["Country/Region"]=="India"]
india
```

```python
#Find the total number of states
india['Province/State'].nunique()
```

```python
#Names of the state
india['Province/State'].unique()
```

```python
#Replace Nan and Unknown by mode

mode_ = india['Province/State'].mode()[0]
india['Province/State'] = india['Province/State'].replace('Unknown', mode_)
india['Province/State'].unique()
```

```python
## India latest Data
india_latest_data=india[india['ObservationDate']==india['ObservationDate'].max()]
india_latest_data
```

```python
## State having maximum cases
top_state=india_latest_data.groupby('Province/State')[["Confirmed",'Recovered']].sum().reset_index()
top_state=top_state.sort_values(['Confirmed'],ascending=False)
top_state.head(20)

```
